```python
import SimpleITK as sitk
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def downloadImage(path):
    reader = sitk.ImageFileReader()
    reader.SetImageIO("NiftiImageIO")
    reader.SetFileName(path)
    image = reader.Execute()
    return image
```

```python
dir_path = '/Users/josephinebregazzi/Work/summer2023/MIPLAB/Brain-Scan-Example-Data/Interrsect/mini-dataset/'
img1_path = 'INTERRSeCT01-005_FollowUp.nii.gz'
img2_path = 'INTERRSeCT01-006_FollowUp_skull_stripped.nii.gz'
img3_path = 'INTERRSeCT01-019_FollowUp_skull_stripped.nii.gz'

image1_3d = downloadImage(dir_path + img1_path)
image2_3d = downloadImage(dir_path + img2_path)
image3_3d = downloadImage(dir_path + img3_path)
```

```python
import os

def saveArray2d_npy(array, out_directory, fname):
    np.save(os.path.join(out_directory, '{}.npy'.format(fname)),
                array,
                allow_pickle=False)
    
    
```

```python
save_to = '/Users/josephinebregazzi/Work/summer2023/MIPLAB/Brain-Scan-Example-Data/Interrsect/mini-dataset-stripped/slice_sample'

arr1 = sitk.GetArrayFromImage(image1)
fname1 = 'intersect_2d_img_1'

arr2 = sitk.GetArrayFromImage(image2)
fname2 = 'intersect_2d_img_2'

arr3 = sitk.GetArrayFromImage(image3)
fname3 = 'intersect_2d_img_3'

saveArray2d_npy(arr1, save_to, fname1)
saveArray2d_npy(arr2, save_to, fname2)
saveArray2d_npy(arr3, save_to, fname3)
```
